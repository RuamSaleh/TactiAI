# Exploratory Data Analysis (EDA)

In this notebook, we will perform exploratory data analysis on the tennis matches dataset to understand the features and the target variable `tactical_decision`. We will visualize the data and derive insights that can help in building a predictive model.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualisation style
sns.set(style='whitegrid')
```

```python
# Load the dataset
data = pd.read_csv('../data/tennis_matches_with_tactics.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='tactical_decision')
plt.title('Distribution of Tactical Decisions')
plt.xlabel('Tactical Decision')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize relationships between features and the target variable
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='tactical_decision', y='minutes')
plt.title('Minutes Played by Tactical Decision')
plt.xlabel('Tactical Decision')
plt.ylabel('Minutes')
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])
```

```python
correlation = numeric_data.corr()
```

```python
# Correlation heatmap
plt.figure(figsize=(14, 10))

# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Calculate correlation
correlation = numeric_data.corr()

# Plot heatmap
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Conclusion

In this notebook, we performed exploratory data analysis on the tennis matches dataset. We visualized the distribution of the target variable and examined relationships between features. These insights will guide us in the data preprocessing and model training phases.
